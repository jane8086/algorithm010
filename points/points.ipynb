```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
image_points = pd.read_csv("points/image_points.csv",header=None,sep=' ')
points_world_pixel = pd.read_csv("points/points_world_pixel.csv",header=None,sep=' ')
points_world = pd.read_csv("points/points_world.csv",header=None,sep=' ')
```

# image_points

```python
ps1 = image_points
```

```python
ps1_col = ps1.columns.size # 2列
ps1_row = ps1.iloc[:,0].size #1241626行
ip = np.zeros_like(ps1)
for i in range(ps1_row):
    temp_row = ps1[i:i+1]
    _list = np.ndarray.tolist(temp_row.values)
    _str = ','.join(str(_list) for s in _list if s not in ['[',']'])
    _split = _str.split()
    for j in range(ps1_col):
        _r = list(filter(lambda ch: ch in '-0123456789e.nan', _split[j]))
        _r = "".join(_r)
        ip[i][j] = _r
```

```python
for i in range(ps1_row):
    for j in range(ps1_col):
        ip[i][j] = float(ip[i][j])
```

```python
plt.scatter(ip[:,0],ip[:,1],color="green")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("image_points")
plt.show()
```

# points_world

```python
ps3 = points_world
```

```python
ps3_col = ps3.columns.size # 3列
ps3_row = ps3.iloc[:,0].size #1241626行
wp = np.zeros_like(ps3)
for i in range(ps3_row):
    temp_row = ps3[i:i+1]
    _list = np.ndarray.tolist(temp_row.values)
    _str = ','.join(str(_list) for s in _list if s not in ['[',']'])
    _split = _str.split()
    for j in range(ps3_col):
        _r = list(filter(lambda ch: ch in '-0123456789e.nan', _split[j]))
        _r = "".join(_r)
        wp[i][j] = _r
```

```python
index = []
for i in range(ps3_row):
    for j in range(ps3_col):
        if wp[i][j]=='-nan':
            index.append(i)
wp_valid = wp[0:index[1],:]
```

```python
for i in range(index[1]):
    for j in range(ps3_col):
        wp_valid[i][j] = float(wp_valid[i][j])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(wp_valid[:,0], wp_valid[:,1], wp_valid[:,2])
plt.title("rebulid_screen")
plt.show()
```
